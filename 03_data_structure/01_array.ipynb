### Tablica (array)

Uwaga! Struktury omawiane w tym rozdziale dotyczą teorii struktur danych i nie odnosimy się w nich jeszcze do konkretnych struktur danych zimplementowanych w pythonie takich jak pythonowe listy czy słowniki.

Tablica to ciągły blok kodu o stałym rozmiarze elementów.

Ciągły?

Jeżeli ma być ciągły to na samym początku musimy znać jej długość, żeby wiedzieć gdzie ją można wstawić, a gdzie się takie coś nie zmieści.

Jakie są plusy i minusy tablic?

| Operacja           | Najgorsza złożoność     |
|--------------------|-------------------------|
| Wstawianie (insert)| O(n)                   | (bo na końcu)
| Usuwanie (delete)  | O(n)                   | (bo na początku i trzeba przesuwać w celu zachowania ciągłości)
| **Dostęp (access)**    | **O(1)**                   |
| Szukanie (search)  | O(n)                   |

Natychmiastowy dostęp, bo wszystkie elementy jeden obok drugiego. Możemy wyliczyć adres n-tego elementu i od razu do niego skoczyć.

Wstawianie $O(n)$, bo w najgorszym przypadku już nie ma miejsca. Usuwanie $O(n)$, bo w najgorszym przypadku usuwamy z początku i trzeba wszystko przesuwać, żeby zachować ciągłość. 

Czy możemy coś poprawić?

Chcielibyśmy mieć lepsze wstawianie i usuwanie. Możemy, ale coś kosztem czegoś. Co się stanie jeżeli zrezygnujemy z warunku ciągłości?
