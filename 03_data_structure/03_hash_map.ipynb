### Dystrybucja hashy

```python
import uuid
import matplotlib.pyplot as plt

hash_map = HashMap(100)
for _ in range(100_000):
    hash_map.put(uuid.uuid4(), 'some_value')
```

```python
# distribution of the buckets' lenght
X = []
y = []

for idx, bucket in enumerate(hash_map.buckets):
    X.append(idx)
    y.append(len(bucket))

plt.bar(X, y);
```
