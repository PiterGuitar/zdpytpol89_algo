## Dwa wskaźniki

Wzorzec dwóch wskaźników zakłada, że masz dwie zmienne reprezentujące przedział. Przesuwasz te zmienne w wybranych kierunkach, w zależności od problemu.

Na przykład, załóżmy, że masz dany ciąg znaków i musisz sprawdzić, czy jest palindromem.

Aby rozwiązać ten problem za pomocą dwóch wskaźników, użyj jednego wskaźnika do iteracji od początku, a drugiego od końca. Na każdym kroku porównuj wartości na obu wskaźnikach. Jeśli się nie zgadzają, ciąg nie jest palindromem. Jeśli się zgadzają, przesuń wskaźniki bliżej siebie. Kontynuuj, aż wskaźniki się spotkają.

```python
def is_palindrome(text):
    low = 0
    hi = len(text) - 1

    while low < hi:
        if text[low] != text[hi]:
            return False
        low += 1
        hi -= 1
    return True

s = "abcddcba"
res = is_palindrome(s)
res
```

Wzorzec dwóch wskaźników może zmniejszyć złożoność czasową wielu problemów z $O(n^2) \rightarrow O(n)$.

##### 167. Two Sum II - Input Array Is Sorted

Brute force solutions

Uwaga! \
**1-indexed array**

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        for idx1, item1 in enumerate(numbers, start=1):  # 1-indexed array
            for idx2, item2 in enumerate(numbers, start=1):
                if item1 + item2 == target and idx1 != idx2:  # nie może być ta sama
                    # bo jak mamy w liście np. 3 a target jest 6, to samo 3 wziete
                    # dwa razy da nam 6
                    return [idx1, idx2]
```

Przy submicie dostaniemy timeout

Jak przyśpieszyć?

Two pointers pattern

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        low_idx = 0
        hi_idx = len(numbers)-1

        while low_idx < hi_idx:
            item = numbers[low_idx] + numbers[hi_idx]
            if item==target:
                return [low_idx+1, hi_idx+1]
            elif item > target:
                hi_idx -= 1
            else:
                low_idx += 1
```
