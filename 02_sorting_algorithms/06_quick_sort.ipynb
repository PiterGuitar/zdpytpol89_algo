```python
def quick_sort(arr: list) -> None:
    """
    Sorting given array in place using isertion merge algorithm.

    :param arr: unsorted array
    :return: None
    """
    arr_len = len(arr)
    # dno rekurencji
    if arr_len <= 1:
        return arr

    # wybr pivota (rodkowy element)
    pivot_idx = arr_len // 2 
    pivot_value = arr[pivot_idx]
    
    # budujemy lew i praw podtablic
    left = []
    right = []
    for idx in range(arr_len):
        if idx != pivot_idx:
            value = arr[idx]
            if value < pivot_value:
                left.append(value)
            else:
                right.append(value)

    output = quick_sort(left) + [pivot_value] + quick_sort(right)
    return output
```

```python
arr = [1, 2, 3, 44, 38, 5, 47, 15, 9]

arr = quick_sort(arr)
print(arr)
```

Czy pivotem musi by warto rodkowa ?

Nie, pivot moe by dowoln liczb. Na przykad zamy, e pivotem jest pierwszy element listy.

```python
def quick_sort(arr: list) -> None:
    """
    Sorting given array in place using isertion merge algorithm.

    :param arr: unsorted array
    :return: None
    """
    arr_len = len(arr)
    # dno rekurencji
    if arr_len <= 1:
        return arr

    # wybr pivota (pierwszy element)
    pivot_idx = 0
    pivot_value = arr[pivot_idx]
    
    # budujemy lew i praw podtablic
    left = []
    right = []
    for idx in range(arr_len):
        if idx != pivot_idx:
            value = arr[idx]
            if value < pivot_value:
                left.append(value)
            else:
                right.append(value)

    output = quick_sort(left) + [pivot_value] + quick_sort(right)
    return output
```

```python
arr = [1, 2, 3, 44, 38, 5, 47, 15, 9]


arr = quick_sort(arr)
print(arr)
```

Czy to jak liczb wybierzemy jako pivot ma z naczenie dla wydajnoci algorytmu?

Okazuje si, e tak. Ale nie tyle pozycja na licie co jego warto. Jeeli warto pivota bdzie bliska medianie elementw listy, to podzia listy na dwie podlisty bdzie rwnomierny i z kadym krokiem bdziemy zmniejszali dugo naszej listy o poow. I wtedy mamy $log{n}$ podziaw listy. Ale jeeli za kadym razem jako pivot bdziemy wybierali najmniejsz lub najwiksz warto z listy to wszystkie pozostae elementy bd ldoway w jednej podlicie. W ten sposb przy kadym podziale dugo naszej listy bdzie malaa tylko o 1. Trzeba bdzie wykona $n$ podziaw i dla kadego podziau wykona $n$ porwna. Tym samym zoono algorytmu wzronie do $n^2$.

Podsumowujc, algorytm **szybkiego sortowania** (zoptymalizowany) charakteryzuje si:
- kwadratow czasow zoonoci obliczeniow w najgorszym przypadku (aka w przypadku pesymistycznym) (ang. worst-case time time complexity) <div style="border: 1px solid;">$$O(n^2)$$</div>
- liniow czasow zoonoci obliczeniow w najlepszym przypadku (aka w przypadku optymistycznym) (ang. best-case complexity) $$\Omega(n\cdot\log{n})$$
- kwadratow czasow zoonoci obliczeniow w przypadku rednim (ang. average-case time complexity) $$\Theta(n\cdot\log{n})$$
- sta zoonoci pamiciow (ang. space complexity) <div style="border: 1px solid;">$$O(n)$$</div>
