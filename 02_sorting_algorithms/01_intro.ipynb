### Klasyfikacja algorytmów sortowania

Algorytmy sortowania dzielimy:

- ze względu na stabliność na:
  - stabilne - podczas sortowania gwarantują zachowanie kolejności wsytąpień elementów o tej samej wartości
  - niestabline - podczas sortowania nie gwaranują zachowania kolejności wystąpień elementów o tej samej wartości
