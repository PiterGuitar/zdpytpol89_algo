### Sortowanie przez wybieranie (ang. selection sort)
https://en.wikipedia.org/wiki/Selection_sort

[Select-sort with Gypsy folk dance](https://www.youtube.com/watch?v=Ns4TPTC8whw)

```python
def selection_sort(arr: list) -> None:
    """
    Sorting given array in place using selection sort algorithm.

    :param arr: unsorted array
    :return: None
    """
    arr_len = len(arr)

    for low_idx in range(arr_len):  # n
        min_idx = low_idx
        for high_idx in range(low_idx+1, arr_len):  # n-1, n-2, ..., 2, 1 
            if arr[min_idx] > arr[high_idx]:
                min_idx = high_idx

        arr[low_idx], arr[min_idx] = arr[min_idx], arr[low_idx]

    return None
```

```python
arr = [10, 21, 11, 123, 24, 44, 654, 17]

selection_sort(arr)
print(arr)
```

Jak wygląda sprawa ze złożonością obliczeniową tego algorytmu ?

- czasowa najgorsza
  $$O(n)=n^2$$
- czasowa najlepsza
  $$\Omega(n)=n^2$$
- czasowa średnia
  $$\Theta(n)=n^2$$
- pamięciowa
  $$\Theta(n)=n^2$$


Podsumowując, algorytm **sortowania przez wybieranie** charakteryzuje się:
- kwadratową czasową złożonością obliczeniową w najgorszym przypadku (aka w przypadku pesymistycznym) (ang. worst-case time time complexity) <div style="border: 1px solid;">$$O(n^2)$$</div>
- liniową czasową złożonością obliczeniową w najlepszym przypadku (aka w przypadku optymistycznym) (ang. best-case complexity) $$\Omega(n^2)$$
- kwadratową czasową złożonością obliczeniową w przypadku średnim (ang. average-case time complexity) $$\Theta(n^2)$$
- stałą złożonością pamięciową (ang. space complexity) <div style="border: 1px solid;">$$O(1)$$</div>
